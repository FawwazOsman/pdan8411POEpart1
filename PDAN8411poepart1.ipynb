Linear Regression Explanation :

The method used in statistics for simulating the relationship between one or more independent variables and a dependent variable is called linear regression. This relationship is assumed to be linear, which means that changes in the independent variables cause a corresponding change in the dependent variable. The objective is to estimate coefficients that minimize the difference between observed and expected values in order to obtain the best-fitting line, also known as a hyperplane, which determines this connection. In a variety of fields, linear regression is commonly used for analyzing and making predictions about the correlations between variables.

Numerous numerical predictor factors, such as "Year of Purchase," "KM's driven," and maybe additional characteristics like engine size, horsepower, etc., are included in the dataset. To anticipate the cost of the autos, a linear regression model can make use of these numerical factors as features.

One way to find out if newer cars are generally more expensive is to investigate the correlation between the year of purchase and the car's price.

Interpretation: Linear regression models provide interpretable coefficients that indicate the strength and direction of the relationship between each predictor variable and the target variable. This can help in understanding the factors that influence the price of used cars, such as the year of purchase, mileage, and other features.

Building Regression Models: You can build strong regression models to precisely estimate car prices using a dataset of over 9000 rows and specific used car characteristics. Split the data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Bringing in the model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

Train the model

```python
model.fit(X_train, y_train)
```

Make predictions

```python
y_pred = model.predict(X_test)
```

Evaluate the model

Making Predictions

```python
predictions = model.predict(X_test)
predictions
```

Determining the mean

```python
df['Price'].mean()
```

Mean Absolute Error

```python
mae=mean_absolute_error(y_test, predictions)
print("Mean Absolute Error (MAE):", mae)
```

Mean Squared Error

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Coefficient of Determination (R^2):", r2)
```

```python
rmse=np.sqrt(mean_squared_error(y_test, predictions))
print("Root Mean Squared Error (RMSE):", rmse)
```
